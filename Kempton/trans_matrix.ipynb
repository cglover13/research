```python
import numpy as np
import scipy.linalg as la
import nb_general as nb
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph([[1,2],[2,3],[3,4],[4,5],[5,3],[3,1],[4,2],[4,6],[6,3]])
nx.draw(G)
```

```python
C, B = nb.to_edge_space(G,graph=False,B=True)
```

```python
D_hat = np.diag(np.sum(C,axis=1))
I = np.eye(D_hat.shape[0])
```

```python
K = nb.create_k(G)
```

```python
np.sort(la.eigvals(B))
```

```python
np.sort(la.eigvals(K))
```

```python
np.sort(la.eigvals(la.inv(D_hat-I)@B))
```

```python
A = nx.adjacency_matrix(G).todense()
D = np.diag(np.array(np.sum(A,axis=1).T)[0])
small_I = np.eye(D.shape[0])
```

```python
np.sort(la.eigvals(np.block([[la.inv(D-small_I)@A,D-small_I],[-small_I,np.zeros_like(D)]])))
```
