```python
import non_backtracking_tools as nbt
import networkx as nx
import new_non_nb_tools as new
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
# Create random d-regular graph
d = np.random.randint(2,30)*2
n = 1000
G = nx.random_regular_graph(d,n)
```

```python
# Find B matrix
B = nbt.edge_adjacency_nb_matrix(G)
B_graph = nx.convert_matrix.from_numpy_matrix(B)
```

```python
# Find C matrix
C = nbt.edge_adjacency_matrix(G)
C_graph = nx.convert_matrix.from_numpy_matrix(C)
```

```python
# Calculate bound value
bound = 2*np.sqrt(d-1)
bound2 = 2*np.sqrt(d-2)
```

```python
# Epsilon value
tol = 2*np.sqrt(d-1)-2*np.sqrt(d-2)
```

```python
# Eigenvalues of B and C
eigs_b = sorted(la.eig(B)[0])[::-1]
eigs_c = sorted(la.eig(C)[0])[::-1]
```

```python
# 2nd eigenvalue of B and C
lam_b = eigs_b[1]
lam_c = eigs_c[1]
```

```python
print(bound2,lam_b,bound2+tol,bound,lam_c)
```

```python
# Plot original graph
nx.draw(G,pos=nx.spring_layout(G))
plt.draw()
plt.title('Original Graph')
plt.show()
```

```python
# Plot C graph
nx.draw(C_graph,pos=nx.spring_layout(C_graph))
plt.draw()
plt.title('C Graph')
plt.show()
```

```python
# Plot B graph
nx.draw(B_graph,pos=nx.spring_layout(B_graph))
plt.draw()
plt.title('B Graph')
plt.show()
```
