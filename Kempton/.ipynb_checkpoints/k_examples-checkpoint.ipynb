```python
import networkx as nx
import numpy as np
import nb_general as nb
import scipy.linalg as la
import matplotlib.pyplot as plt
```

# Example of K NOT being diagonalizable with a stationary distribution

```python
G = nx.random_regular_graph(4,6)
nx.draw(G,pos=nx.spring_layout(G))
```

```python
counter_1 = nb.NBEigVals(G)
```

## Test for diagonalizability

```python
np.allclose(counter_1.K,la.solve(counter_1.kvecs,np.eye(len(counter_1.mu_vals)))@np.diag(counter_1.mu_vals)@counter_1.kvecs)
```

```python
print(counter_1.K)
```

```python
print(la.solve(counter_1.kvecs,np.eye(len(counter_1.mu_vals)))@np.diag(counter_1.mu_vals)@counter_1.kvecs)
```
