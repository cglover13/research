# Example of non-backtracking with GCD of cycle length =1

```python
g = nx.cycle_graph(3)
C, B = nb.to_edge_space(g,B=True,graph=False)
D = np.diag(np.sum(C,axis=1))
I = np.eye(D.shape[0])
A = np.array(nx.adjacency_matrix(g).todense())
P = la.inv(np.diag(np.sum(A,axis=1)))@A
```

```python
print(la.eigvals(la.inv(D-I)@B))
print(la.eigvals(B))
print(la.eigvals(A))
print(la.eigvals(P))
```
