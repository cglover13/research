```python
G = nx.gnp_random_graph(15,.7)
```

```python
ex_1 = nb.NBEigVals(G)
```

```python
vec1 = np.array([-1 for i in range(ex_1.n)] + [1 for i in range(ex_1.n)])
```

```python
D = np.diag(vals)
```
