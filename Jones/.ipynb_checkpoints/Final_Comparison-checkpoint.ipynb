# Single Cut: Non-backtracking vs Backtracking

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import os
import classes.nb_general as NB_G 
import classes.visual_tools as VS
import scipy.linalg as la
```

```python
def get_single_cut(k, eps=.9):
    
    G = nx.complete_graph(k)
    G.remove_edge(0,1)
    G_ = G
    G = nx.to_numpy_array(G)
    G /= np.sum(G,axis=1)
    Q = np.eye(G.shape[0])-eps*G
    
    print(Q)
```

```python
eps = .9

for i in range(4, 13):
    K = nx.complete_graph(i)
    K.remove_edge(0,1)
    _, A = NB_G.to_edge_space(K, B=True)
    
    G = nx.to_numpy_array(A)
    G /= np.sum(G,axis=1)
    Q_e = np.eye(G.shape[0])-eps*G
    
    plt.subplot(121)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(Q_e, cmap='gray')
    #plt.subplot(132)
    inv_ = la.inv(Q_e)
    #plt.scatter(np.arange(0, len(inv_.flatten()), 1), inv_.flatten())
    plt.subplot(122)
    plt.imshow(inv_, cmap='gray')
    plt.show()
    print("K: {}, Inv Vals: {}".format(i, len(np.unique(inv_.flatten()))))
```
