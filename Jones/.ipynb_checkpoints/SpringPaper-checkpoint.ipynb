```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = [1,2,3,4,5]
E = [(1,4),(1,5),(2,4),(2,5),(3,4),(3,5)]

G = nx.Graph()
G.add_nodes_from(V)
G.add_edges_from(E)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
A = nx.to_numpy_array(G)
A
```
